```python
recs_dir = 'data'
packets_concated = bytes(0)
for f in os.listdir('data')[:1]:
    with open(os.path.join('data',f), 'rb') as rec:
        packets_concated = packets_concated + rec.read()
```

```python
packets = [packets_concated[1400*i:1400*i+1400] for i in range(int(len(packets_concated)/1400))]
```
